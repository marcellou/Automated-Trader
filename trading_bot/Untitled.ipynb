```python
import quandl 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
amzn = quandl.get('WIKI/AMZN', start_date='2017-01-01', end_date='2018-01-01')
```

```python
short_window = 30
long_window = 90
signals = pd.DataFrame(index=amzn.index)
signals['signal'] = 0.0
signals['short_mavg'] = amzn['Close'].rolling(window=short_window, min_periods=1, center=False).mean()
signals['long_mavg'] = amzn['Close'].rolling(window=long_window, min_periods=1, center=False).mean()
signals['signal'][short_window:] = np.where(signals['short_mavg'][short_window:]
                                            > signals['long_mavg'][short_window:], 1.0, 0.0)
signals['positions'] = signals['signal'].diff()
```

```python
initial_capital = float(50000.0)
positions = pd.DataFrame(index=signals.index).fillna(0.0) #using the signals dataframe
positions['AMZN'] = 50*signals['signal']
#on the days the the signal is equal to 1 and the SMA crosses the LMA  / BUY 50 shares
portfolio = positions.multiply(amzn['Adj. Close'], axis=1)
pos_diff = positions.diff()
portfolio['holdings'] = (positions.multiply(amzn['Adj. Close'], axis=0)).sum(axis=1)
portfolio['cash'] = initial_capital - (pos_diff.multiply(amzn['Adj. Close'], axis=0)).sum(axis=1).cumsum()
portfolio['total'] = portfolio['cash'] + portfolio['holdings']
portfolio['returns'] = portfolio['total'].pct_change()
```

```python
fig = plt.figure(figsize=(14,8))
ax1 = fig.add_subplot(111, ylabel='Portfolio value in $')
portfolio['total'].plot(ax=ax1)
ax1.plot(portfolio.loc[signals.positions == 1.0].index,
        portfolio.total[signals.positions == 1.0],
        '^', markersize=10, color='m')
ax1.plot(portfolio.loc[signals.positions == -1.0].index,
        portfolio.total[signals.positions == -1.0],
        'v', markersize=10, color='k')
plt.show()

#signals 1.0 up arrow
#signals that have negative yield = -1.0
#sharpe ratio implenentation
```

```python
returns = portfolio['returns']
np.sqrt(252) * (returns.mean() / returns.std())
#sharpe ratio
#ratio grater than 1 is traditionally good..

#developed and automaded stock trader module
#implemented stock market analysis
#designed a trading algorithm
#structure a trading bot and ran trades on it..
```
